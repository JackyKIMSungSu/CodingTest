```python
n = int(input())

data = input(f'숫자 {n}개를 입력하세요. :')
tokens = data.split()
h = [int(token) for token in tokens]

ah = h[0]
ac = 0
remains = 0

while remains > (-1*n):
  for i in range(n):
    if h[i] == ah:
      h[i] = -1
      ah -= 1
      
    remains += h[i]
    if ah == 0:
      break
    
  
  ac += 1
  if remains == (-1*n):
    break
  remains = 0

  for i in range(n):
    if h[i] > -1:
      ah = h[i]
      break
  

print(ac)
```
